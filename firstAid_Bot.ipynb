```python
from gtts import gTTS
import base64

def play_audio(text):
    # Clean text of markdown characters for better speech
    clean_text = text.replace('*', '').replace('#', '')
    tts = gTTS(text=clean_text, lang='en')
    tts.save('instructions.mp3')

    with open('instructions.mp3', 'rb') as f:
        data = f.read()
        b64 = base64.b64encode(data).decode()
        md = f'<audio controls autoplay><source src="data:audio/mp3;base64,{b64}" type="audio/mp3"></audio>'
        display(widgets.HTML(md))
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, Markdown
from datetime import datetime

# Ensure global state for audio
if 'last_response' not in globals():
    last_response = ""

speak_button = widgets.Button(
    description='READ ALOUD',
    button_style='info',
    icon='volume-up',
    layout=widgets.Layout(width='150px', height='40px')
)

def on_speak_clicked(b):
    global last_response
    if last_response:
        with output:
            play_audio(last_response)
    else:
        with output:
            print("No instructions available to read. Please generate them first.")

speak_button.on_click(on_speak_clicked)

def on_get_clicked_updated(b):
    global last_response
    user_input = input_text.value.strip()

    with output:
        clear_output(wait=True)
        if not user_input:
            display(Markdown("**⚠️ Please enter an injury description first.**"))
            return

        current_time = datetime.now().strftime("%I:%M %p")
        display(Markdown(f"⏳ *Analyzing situation at {current_time}...*"))

        try:
            # Fetch instructions from the initialized agent
            last_response = agent.get_instructions(user_input)
            clear_output(wait=True)
            display(Markdown(f"### 📋 Results for: {user_input.capitalize()}"))
            display(Markdown(f"*Generated at {current_time}* flock"))
            display(Markdown("---"))
            display(Markdown(last_response))
        except Exception as e:
            display(Markdown(f"**⚠️ Error generating instructions:** {str(e)}"))

# Reset and re-attach handlers to avoid multiple triggers
get_button.on_click(on_get_clicked_updated, remove=True)
get_button.on_click(on_get_clicked_updated)

# Re-display the UI
buttons = widgets.HBox([get_button, speak_button, clear_button], layout=widgets.Layout(gap='10px', margin='10px 0'))
ui_container = widgets.VBox([ui_header, input_text, buttons, output])
display(ui_container)
```

### Speech-to-Text Integration
The following code adds a JavaScript bridge to record audio directly in the browser and transcribe it into the input field.

```python
protocol_history = []

def update_history_display():
    with history_output:
        clear_output(wait=True)
        if not protocol_history:
            display(Markdown("No history yet. Start by generating a protocol!"))
            return
        history_md = "<div style='max-height: 300px; overflow-y: auto;'>"
        for entry in reversed(protocol_history):
            history_md += f"<div style='border-bottom: 1px dashed #dcdde1; padding: 8px 0; font-size: 0.9em;'>"
            history_md += f"<b style='color:#7f8c8d;'>{entry['time']}</b>: {entry['query']}"
            history_md += "</div>"
        history_md += "</div>"
        display(Markdown(history_md))

def on_get_clicked_final(b):
    global last_response
    user_input = input_text.value.strip()
    if not user_input: return

    with output:
        clear_output(wait=True)
        current_time = datetime.now().strftime("%H:%M:%S")
        display(widgets.HTML("<div style='color:#1abc9c'>📡 Establishing AI Connection...</div>"))
        try:
            full_prompt = f"Provide first aid for: {user_input}. Format your response exactly like this: [INSTANT RELIEF: One short sentence to stop pain/damage] followed by [FULL PROTOCOL: detailed steps]."
            completion = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[
                    {"role": "system", "content": agent.system_prompt},
                    {"role": "user", "content": full_prompt}
                ],
                temperature=0.3
            )
            raw_response = completion.choices[0].message.content
            last_response = raw_response

            relief_text = "Stay calm, monitor breathing, and call emergency services."
            main_protocol = raw_response

            if "[INSTANT RELIEF:" in raw_response and "[FULL PROTOCOL:" in raw_response:
                try:
                    relief_text = raw_response.split("[INSTANT RELIEF:")[1].split("]")[0]
                    main_protocol = raw_response.split("[FULL PROTOCOL:")[1].split("]")[0]
                except:
                    pass

            clear_output(wait=True)
            display(widgets.HTML(f"""
                <div class='relief-box'>
                    <b style='color:#1b5e20; font-size:1.1em;'>⚡ INSTANT REMEDY / REDUCE PAIN:</b><br>
                    <span style='color:#2e7d32; font-size:1.05em;'>{relief_text}</span>
                </div>
                <div class='clinical-report'>
                    <div style='color:#7f8c8d; font-size:0.8em; margin-bottom:10px;'>REPORT ID: {current_time} | STATUS: ACTIVE</div>
                    <div style='font-size:1.2em; font-weight:700; color:#2c3e50; margin-bottom:15px;'>ANALYSIS: {user_input.upper()}</div>
                    <div style='line-height:1.6; color:#2f3640;'>{main_protocol.replace('\n', '<br>')}</div>
                </div>
            """))
            protocol_history.append({'time': current_time, 'query': user_input.capitalize()})
            update_history_display()
        except Exception as e:
            display(widgets.HTML(f"<div class='persistent-warning'>SYSTEM ERROR: {str(e)}</div>"))

get_button.on_click(on_get_clicked_final, remove=True)
get_button.on_click(on_get_clicked_final)

layout_grid = widgets.VBox([
    ui_header,
    widgets.VBox([
        input_text,
        widgets.HBox([get_button, mic_button, speak_button, clear_button, clear_hist_button],
                     layout=widgets.Layout(justify_content='center', gap='10px', margin='10px 0'))
    ], layout=widgets.Layout(padding='0 20px')),
    widgets.HBox([
        widgets.VBox([widgets.HTML("<h4 style='color:#2c3e50; margin-left:20px;'>📋 CLINICAL PROTOCOL</h4>"), output], layout=widgets.Layout(width='65%')),
        widgets.VBox([widgets.HTML("<h4 style='color:#2c3e50;'>📜 SESSION LOG</h4>"), history_output], layout=widgets.Layout(width='35%'))
    ], layout=widgets.Layout(padding='20px', gap='20px'))
], layout=widgets.Layout(background_color='#f0f4f8', border_radius='20px', border='1px solid #dcdde1'))

display(layout_grid)
update_history_display()
```
